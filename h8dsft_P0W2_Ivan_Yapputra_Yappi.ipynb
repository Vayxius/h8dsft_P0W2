## Data Retrieve (Query From Google Cloud)

Pada bagian ini hanya menempelkan salinan query yang sudah dibuat sebelumnya di Google Cloud Big Query.

```  mysql
SELECT trips.trip_id, trips.start_date, trips.start_station_name, trips.end_station_name, trips.start_station_latitude, trips.start_station_longitude, trips.end_station_latitude, trips.end_station_longitude, info.region_id

FROM `bigquery-public-data.san_francisco_bikeshare.bikeshare_trips` trips 

JOIN `bigquery-public-data.san_francisco_bikeshare.bikeshare_station_info` info ON trips.start_station_name = info.name

WHERE start_station_name != end_station_name AND start_station_latitude is not null AND start_station_longitude is not null AND end_station_latitude is not null AND end_station_longitude is not null AND start_date between '2017-11-01' AND '2018-04-01'
```

- Buatlah dua variable baru yang bernama `start_stations` dan `end_stations` yang masing-masing berisikan list nama stasiun awal dan stasiun akhir. Pastikan tidak ada nama stasiun yang duplikat di masing-masing variable.

````python
# Memasukkan semua list nama stasiun awal di dalam dataframe bike_system ke dalam variable start_stations
# Method .drop_duplicates() berfungsi untuk menghapus nama yang sama di dalam kolom start_station_name
start_stations = bike_system['start_station_name'].drop_duplicates()

# Memasukkan semua list nama stasiun akhir di dalam dataframe bike_system ke dalam variable end_stations
# Method .drop_duplicates() berfungsi untuk menghapus nama yang sama di dalam kolom end_station_name
end_stations = bike_system['end_station_name'].drop_duplicates()
````

Memasukkan semua data pada kolom `start_station_name` ke dalam variable `start_stations`, dan semua data pada kolom `end_station_name` ke dalam variable `end_stations`, dimana jika di dalam kolom tersebut terdapat nama yang sama, maka hanya 1 data yang akan dimasukkan kedalam variable `start_stations` atau `end_stations`.

- Deteksi stasiun-stasiun yang hanya terdapat di salah satu variable dan tidak di keduanya (exclusive-or/xor) menggunakan np.setxor1d(array1,array2). Outputnya akan berupa numpy array dan masukkan output tersebut ke dalam variable bernama `exclusive_stations`.

````python
exclusive_stations = np.setxor1d(start_stations,end_stations)
````
